```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.cricbuzz.com/profiles/265/ms-dhoni"
page=requests.get(url)
soup=BeautifulSoup(page.text,'lxml')
```

```python
table=soup.find('table',class_="table cb-col-100 cb-plyr-thead")
#table

headers=[]

for i in table.find_all('th'):
    title=i.text
    headers.append(title)
#print(headers)

df1=pd.DataFrame(columns=headers)
#df

for rows in table.find_all('tr')[1:]:
    data=rows.find_all('td')
    rows_data=[td.text for td in data]
    length=len(df1)
    df1.loc[length]=rows_data
    
    
    
        
```

```python
df1
```

```python
df1.to_csv('msd_stats.csv')
```
